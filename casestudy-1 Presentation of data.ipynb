# 1. Read the dataset to the python environment.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('cars_data.csv')
data
                   
```

# 2. Check for the null values present in the dataset.

```python
data.isnull().sum()
```

# no null values are present

# 3.  Plot a bar graph of male vs female buyers participated in the sales.

```python
sns.set(font_scale=1.4)
data['Buyer Gender'].value_counts().plot(kind='bar',figsize=(9,6),rot=0,color=['blue', 'orange'])
plt.xlabel('Gender',labelpad=14)
plt.ylabel('Count of buyers',labelpad=14)
plt.title('Male vs female buyers participated in the sales')
```

# 4. Find the top 5 cars based on their sales price.


```python
data1 = data.nlargest(5,['Sale Price'])
data1
```

```python
data1[['Model','Sale Price']]
```

# Top 5 cars based on sale price are jettaIII, A6, LX, MX-6 and ES.
